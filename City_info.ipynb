# Zillow GeoJSON

```python
import json
import pandas as pd
from pandas.io.json import json_normalize

with open('zillow-neighborhoods.geojson') as f:
    data = json.load(f)
    
df = []

for feature in data['features']:
    df.append(feature["properties"])
    
df = json_normalize(df)

df
```

```python
san_diego = df.loc[(df.city == "San Diego") & (df.state == "CA")]

san_diego.reset_index(inplace = True)

san_diego
```

```python
san_diego[['latitude','longitude']] = pd.DataFrame(san_diego.geo_point_2d.tolist(), index = san_diego.index)
```

```python
san_diego
```

```python
san_diego = san_diego[["name", "city", "state", "latitude", "longitude"]]

san_diego.rename(columns={"name": "Neighborhood", "city": "City", "state": "State", "latitude": "Latitude", "longitude": "Longitude"}, inplace = True)

san_diego
```

```python
def city_info(city, state):
    
    import json
    import pandas as pd

    with open('zillow-neighborhoods.geojson') as f:
        data = json.load(f)

    df = []

    for feature in data['features']:
        df.append(feature["properties"])

    df = json_normalize(df)
       
    df = df.loc[(df.city == city) & (df.state == state)]

    df.reset_index(inplace = True)

    df[['latitude','longitude']] = pd.DataFrame(df.geo_point_2d.tolist(), index = df.index)

    df = df[["name", "city", "state", "latitude", "longitude"]]

    df.rename(columns={"name": "Neighborhood", "city": "City", "state": "State", "latitude": "Latitude", "longitude": "Longitude"}, inplace = True)
    
    return df
```

# Scrape "The Richest Neighborhoods In San Diego For 2020"

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.homesnacks.net/richest-neighborhoods-in-san-diego-129013/"

response_object = requests.get(url).text
soup = BeautifulSoup(response_object, "lxml")
```

```python
richest_sd = soup.find('table', attrs = {"id": "ranking-table"})
richest_sd_data = richest_sd.tbody.find_all("tr")

richest_sd_data
```

```python
# Get all the Headers and Data

headers = []
table_data = []

for th in richest_sd.thead.tr.find_all("th"):
    headers.append(th.text)

for tr in richest_sd_data:
    td = tr.find_all('td')
    row = [i.text for i in td]
    table_data.append(row)
    
richest_sd_df = pd.DataFrame.from_records(table_data, columns = headers)

richest_sd_df['Median Household Income<'] = richest_sd_df['Median Household Income<'].str.replace(',', '').str.replace('$', '').astype(int)

richest_sd_df["Rank"] = richest_sd_df["Rank"].astype('int32')

richest_sd_df.rename(columns={"Median Household Income<": "Median Household Income"}, inplace = True)

print(richest_sd_df.dtypes)

richest_sd_df
```

```python
merged_san_diego = pd.merge(left = richest_sd_df, right=san_diego, left_on='Neighborhood', right_on='Neighborhood')
print(merged_san_diego.shape)
merged_san_diego
```

```python
san_diego_neighborhoods = merged_san_diego[["Neighborhood", "City", "State", "Median Household Income", "Latitude", "Longitude"]]
san_diego_neighborhoods
```

```python
print(san_diego_neighborhoods.dtypes)
```

# Schools in San Diego Neighborhoods in radius of 2000 meters (1.25 miles)

```python
def getNearbySchools(names, latitudes, longitudes, radius=2000):
    
    venues_list=[]
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
            
        # create the API request URL
        
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}&categoryId=4bf58dd8d48988d13b941735'.format(
            CLIENT_ID, 
            CLIENT_SECRET,  
            lat, 
            lng,
            VERSION,
            radius, 
            LIMIT)
            
        # make the GET request
        
        results = requests.get(url).json()["response"]["venues"]
        
        # return only relevant information for each nearby venue
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['name'], 
            v['location']['lat'], 
            v['location']['lng'],  
            v['categories'][0]['name']) for v in results])

    nearby_schools = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    
    nearby_schools.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'School Name', 
                  'School Latitude', 
                  'School Longitude', 
                  'School Category']
    
    return(nearby_schools)
```

```python
def cleaning_schools(df):
    
    df = df.loc[(df["School Category"] == "Elementary School") | (df["School Category"] == "Middle School") | (df["School Category"] == "High School")]
    df.drop_duplicates(subset=['School Latitude', "School Longitude"], keep="first", inplace = True)
    df = df.reset_index(drop = True)
    
    
    return df
```
