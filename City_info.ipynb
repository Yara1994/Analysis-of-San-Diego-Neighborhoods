# Zillow GeoJSON

## I found this GeoJSON file with information about all cities and its neighborhoods in USA on this web page:
### https://data.opendatasoft.com/explore/dataset/zillow-neighborhoods%40public/information/.

## You are welcome to download through this web site or through my github page.

```python
# Import libraries

import json
import pandas as pd
from pandas.io.json import json_normalize

# open Zillow GeoJson file

with open('zillow-neighborhoods.geojson') as f:
    data = json.load(f)

# Create an empty list to store information

df = []

for feature in data['features']:
    df.append(feature["properties"])

# json_normalizer creates dataframe from json format

df = json_normalize(df)

df
```

```python
# Filtering to get all Neighborhoods from San Diego city (NO COUNTY!)

san_diego = df.loc[(df.city == "San Diego") & (df.state == "CA")]

# Reset index

san_diego.reset_index(inplace = True)

san_diego
```

```python
# Create separate columns from a list

san_diego[['latitude','longitude']] = pd.DataFrame(san_diego.geo_point_2d.tolist(), index = san_diego.index)

san_diego
```

```python
# Pick columns that we are interested in

san_diego = san_diego[["name", "city", "state", "latitude", "longitude"]]

# Rename the columns

san_diego.rename(columns = {"name": "Neighborhood", "city": "City", "state": "State", "latitude": "Latitude", "longitude": "Longitude"}, inplace = True)

san_diego
```

## I've created this function to find neighborhoods from the citie you want from in one line of code.


### Just type:

### city_info("City you are interested in", "State of this city")




```python
def city_info(city, state):
    
    import json
    import pandas as pd

    with open('zillow-neighborhoods.geojson') as f:
        data = json.load(f)

    df = []

    for feature in data['features']:
        df.append(feature["properties"])

    df = json_normalize(df)
       
    df = df.loc[(df.city == city) & (df.state == state)]

    df.reset_index(inplace = True)

    df[['latitude','longitude']] = pd.DataFrame(df.geo_point_2d.tolist(), index = df.index)

    df = df[["name", "city", "state", "latitude", "longitude"]]

    df.rename(columns={"name": "Neighborhood", "city": "City", "state": "State", "latitude": "Latitude", "longitude": "Longitude"}, inplace = True)
    
    return df
```

# Scrape "The Richest Neighborhoods In San Diego For 2020"

```python
# Import libraries for scraping

import requests
from bs4 import BeautifulSoup

url = "https://www.homesnacks.net/richest-neighborhoods-in-san-diego-129013/"

response_object = requests.get(url).text
soup = BeautifulSoup(response_object, "lxml")
```

```python
richest_sd = soup.find('table', attrs = {"id": "ranking-table"})
richest_sd_data = richest_sd.tbody.find_all("tr")

richest_sd_data
```

```python
# Get all the Headers and Data

headers = []
table_data = []

# Collect Headers

for th in richest_sd.thead.tr.find_all("th"):
    headers.append(th.text)

# Collect Data

for tr in richest_sd_data:
    td = tr.find_all('td')
    row = [i.text for i in td]
    table_data.append(row)

richest_sd_df = pd.DataFrame.from_records(table_data, columns = headers)

# Change data type of Median Household Income and Rank columns

richest_sd_df['Median Household Income<'] = richest_sd_df['Median Household Income<'].str.replace(',', '').str.replace('$', '').astype(int)
richest_sd_df["Rank"] = richest_sd_df["Rank"].astype('int32')

# Rename columns

richest_sd_df.rename(columns={"Median Household Income<": "Median Household Income"}, inplace = True)

# Making sure all data in correct type

print(richest_sd_df.dtypes)

richest_sd_df
```

```python
# Merge 2 dataframes in one

merged_san_diego = pd.merge(left = richest_sd_df, right=san_diego, left_on='Neighborhood', right_on='Neighborhood')

# Check the shape of dataframe

print(merged_san_diego.shape)

merged_san_diego
```

# Schools in San Diego Neighborhoods in radius of 2000 meters (1.25 miles)

```python
def getNearbySchools(names, latitudes, longitudes, radius=2000, LIMIT = 200):
    
    school_list=[]
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
            
        # create the API request URL
        
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}&categoryId=4bf58dd8d48988d13b941735'.format(
            CLIENT_ID, 
            CLIENT_SECRET,  
            lat, 
            lng,
            VERSION,
            radius, 
            LIMIT)
            
        # make the GET request
        
        results = requests.get(url).json()["response"]["venues"]
        
        # return only relevant information for each nearby school
        
        school_list.append([(
            name, 
            lat, 
            lng, 
            s['name'], 
            s['location']['lat'], 
            s['location']['lng'],  
            s['categories'][0]['name']) for s in results])

    nearby_schools = pd.DataFrame([item for school_list in school_list for item in school_list])
    
    nearby_schools.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'School Name', 
                  'School Latitude', 
                  'School Longitude', 
                  'School Category']
    
    return(nearby_schools)
```

```python
# Define a Function to select the top 3 rows (According to the Values) 

def return_most_common_schools(row, num_top_schools):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending = False)
    
    return row_categories_sorted.index.values[0:num_top_schools]


num_top_schools = 3

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues

columns = ['Neighborhood']

for ind in np.arange(num_top_schools):
    try:
        columns.append('{}{} Most Common School'.format(ind + 1, indicators[ind]))
    except:
        columns.append('{}th Most Common School'.format(ind + 1))

# create a new dataframe

richest_top_3_sorted = pd.DataFrame(columns = columns)
richest_top_3_sorted['Neighborhood'] = richest_neighborhoods_onehot_grouped['Neighborhood']

for ind in np.arange(richest_neighborhoods_onehot_grouped.shape[0]):
    richest_top_3_sorted.iloc[ind, 1:] = return_most_common_schools(richest_neighborhoods_onehot_grouped.iloc[ind, :], num_top_schools)

    
richest_top_3_sorted.index = np.arange(1, len(richest_top_3_sorted) + 1)

richest_top_3_sorted
```
